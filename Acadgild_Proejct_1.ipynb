```python
# Q12 Join users to transactions, keeping all rows from transactions and only matching rows
# from users (left join)
```

```python
pd.merge(transactions,users,how='left') #using left join 
```

```python
# Q13. Which transactions have a UserID not in users?
```

```python
#This can be solved by using the  left outer join as it will give nan for all non matching rows from the user table  
# or better approach is to use isin method
```

```python
# 1 st approach 
# pd.merge(transactions,users,how='left') #using left join 
# or 2nd Approach
transactions[~transactions['UserID'].isin(users['UserID'])]
```

```python
#Q 14 Join users to transactions, keeping only rows from transactions and users that match
# via UserID (inner join) Expected Output:
```

```python
pd.merge(transactions,users,how='inner') #inner join
```

```python
# Q 15 Join users to transactions, displaying all matching rows AND all non-matching rows
# (full outer join)
```

```python
transactions.merge(users,how='outer')#outer join
```

```python
# Q 16 Determine which sessions occurred on the same day each user registered
```

```python
pd.merge(sessions, users,left_on=['UserID','SessionDate'],right_on=['UserID','Registered'])
```

```python
# Q 17 Build a dataset with every possible (UserID, ProductID) pair(Cross Join)
```

```python
users['key']=1
products['key']=1
pd.merge(users,products,on='key',how='inner')[['UserID','ProductID']]
```

```python
# Q 18. Determine how much quantity of each product was purchased by each user.
```

```python
pd.merge(users,transactions , how='inner').groupby(['UserID','ProductID']).sum()  #Grouping o\pn the basis of the user and customer
```

```python
# Q 19 For each user, get each possible pair of pair transactions (TransactionID1,
# TransacationID2)
```

```python
pd.merge(transactions, transactions,left_on='TransactionID',right_on='TransactionID') # self join 
```

```python
# Q 20. Join each user to his/her first occuring transaction in the transactions table
```

```python
#using group by and than applyig the aggregation function first.
pd.merge(users, transactions.groupby('UserID').first().reset_index(),on='UserID') 
```
